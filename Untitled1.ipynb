```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt


# Step 1: Stratified Sampling
df['stratify_column'] = df['SHOT_TYPE'].astype(str) + '_' + df['QUARTER'].astype(str)

# Perform stratified sampling
sample_size = 50000  # Adjust sample size as needed
sampled_data, _ = train_test_split(
    df,
    test_size=(1 - sample_size / len(df)),
    stratify=df['stratify_column'],
    random_state=42
)

# Step 2: Preprocessing - Scaling and PCA
data_for_clustering = sampled_data[selected_features]

# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_for_clustering)

# Reduce dimensions using PCA
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)

# Step 3: DBSCAN Clustering
dbscan = DBSCAN(eps=0.5, min_samples=10)  # Adjust eps and min_samples for tuning
dbscan_labels = dbscan.fit_predict(data_pca)

# Silhouette score for DBSCAN (ignoring outliers labeled as -1)
if len(set(dbscan_labels)) > 1:  # Avoid errors if DBSCAN creates no clusters
    dbscan_silhouette = silhouette_score(data_pca[dbscan_labels != -1], dbscan_labels[dbscan_labels != -1])
    print(f"Silhouette Score for DBSCAN: {dbscan_silhouette}")
else:
    dbscan_silhouette = -1
    print("DBSCAN did not form valid clusters.")

# Visualize DBSCAN Clusters
plt.figure(figsize=(8, 6))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=dbscan_labels, cmap='viridis', s=5)
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('DBSCAN Cluster Visualization')
plt.colorbar(label='Cluster Label')
plt.show()

# Step 4: Agglomerative Clustering
agglomerative = AgglomerativeClustering(n_clusters=3, linkage='ward')  # Test different n_clusters
agglomerative_labels = agglomerative.fit_predict(data_pca)

# Silhouette score for Agglomerative Clustering
agglomerative_silhouette = silhouette_score(data_pca, agglomerative_labels)
print(f"Silhouette Score for Agglomerative Clustering: {agglomerative_silhouette}")

# Visualize Agglomerative Clusters
plt.figure(figsize=(8, 6))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=agglomerative_labels, cmap='viridis', s=5)
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('Agglomerative Clustering Visualization')
plt.colorbar(label='Cluster Label')
plt.show()

# Step 5: Compare Results
print("\n--- Clustering Results Summary ---")
print(f"Silhouette Score for DBSCAN: {dbscan_silhouette}")
print(f"Silhouette Score for Agglomerative Clustering: {agglomerative_silhouette}")
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
from joblib import Parallel, delayed
import matplotlib.pyplot as plt

# Silhouette score calculation function
def compute_silhouette(k, data):
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(data)
    score = silhouette_score(data, cluster_labels)
    return k, score

# Define range of k values for evaluation
k_values = range(2, 10)

# Use parallel processing to compute silhouette scores for all k values
results = Parallel(n_jobs=-1)(delayed(compute_silhouette)(k, player_cluster_distribution) for k in k_values)

# Display silhouette scores
for k, score in results:
    print(f"Silhouette Score for k={k}: {score}")

# Plot silhouette scores
k_values, scores = zip(*results)
plt.plot(k_values, scores, marker='o')
plt.title('Silhouette Scores for Different k Values (Player Data)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid()
plt.show()

plt.show()
```

```python
wcss = []
k_values = range(2, 10)

# Compute WCSS for each k
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(player_cluster_distribution)
    wcss.append(kmeans.inertia_)

# Plot the WCSS for each k
plt.plot(k_values, wcss, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.grid()
plt.show()
```

```python
from sklearn.metrics import calinski_harabasz_score

# Compute Calinski-Harabasz Index for each k
ch_scores = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(player_cluster_distribution)
    ch_score = calinski_harabasz_score(player_cluster_distribution, cluster_labels)
    ch_scores.append(ch_score)

# Plot the Calinski-Harabasz scores
plt.plot(k_values, ch_scores, marker='o')
plt.title('Calinski-Harabasz Index for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Calinski-Harabasz Index')
plt.grid()
plt.show()

# Print Calinski-Harabasz scores
for k, score in zip(k_values, ch_scores):
    print(f"Calinski-Harabasz Index for k={k}: {score}")
```

```python
import numpy as np
import pandas as pd


def compute_gap_stat(data, n_refs=10, max_k=10):
    gaps = []
    for k in range(2, max_k):
        # Fit KMeans to the data
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(data)
        ref_disps = []

        # Generate random reference datasets
        for _ in range(n_refs):
            random_data = np.random.uniform(data.min(axis=0), data.max(axis=0), data.shape)
            kmeans_ref = KMeans(n_clusters=k, random_state=42)
            kmeans_ref.fit(random_data)
            ref_disps.append(kmeans_ref.inertia_)

        # Compute the gap statistic
        gap = np.log(np.mean(ref_disps)) - np.log(kmeans.inertia_)
        gaps.append(gap)

    return gaps

# Compute and plot the gap statistic
gaps = compute_gap_stat(player_cluster_distribution.values, max_k=10)
plt.plot(range(2, len(gaps) + 2), gaps, marker='o')
plt.title('Gap Statistic for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Gap Statistic')
plt.grid()
plt.show()
```

```python
# 1. Calculate 3P_RATE (Proportion of 3-point attempts per team)
three_point_attempts = df[df['SHOT_TYPE'] == '3PT Field Goal'].groupby('TEAM_NAME').size()
total_attempts = df.groupby('TEAM_NAME').size()
team_analysis['3P_RATE'] = (three_point_attempts / total_attempts).fillna(0)

# 2. Calculate RA_RATE (Proportion of shots in the restricted area per team)
restricted_area_attempts = df[df['BASIC_ZONE'] == 'Restricted Area'].groupby('TEAM_NAME').size()
team_analysis['RA_RATE'] = (restricted_area_attempts / total_attempts).fillna(0)

# 3. Calculate EFG% (Effective field goal percentage per team)
# Precompute ThreePM before grouping
df['ThreePM'] = (df['SHOT_TYPE'] == '3PT Field Goal') & (df['SHOT_MADE'] == 1)

team_efg_data = df.groupby('TEAM_NAME').agg(
    FGM=('FGM', 'sum'),
    FGA=('FGA', 'sum'),
    ThreePM=('ThreePM', 'sum')
)
team_efg_data['EFG%'] = (team_efg_data['FGM'] + 0.5 * team_efg_data['ThreePM']) / team_efg_data['FGA']
team_analysis = team_analysis.merge(team_efg_data[['EFG%']], on='TEAM_NAME', how='left')

# Verify the computed metrics
print(team_analysis[['TEAM_NAME', '3P_RATE', 'RA_RATE', 'EFG%']].head())
```
